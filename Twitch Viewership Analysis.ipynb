```python
viewers = np.array([2700, 3400, 4000, 7000, 7800, 10500, 1100, 800, 500, 444, 330, 806, 40, 15, 8, 3, 1])
followers = np.array([83900, 168000, 144000, 231000, 308000, 151000, 20300, 26800, 26500, 79900, 46700, 31500, 2300, 1900, 307, 93, 17])
```

```python
reg = LinearRegression(fit_intercept = False).fit(followers.reshape(-1, 1),viewers)
plt.scatter(followers,viewers)
print("Viewers = " + str(reg.coef_[0]) + "*followers")
plt.plot(np.arange(0,300000), reg.predict(np.arange(0,300000).reshape(-1, 1)), c="r")
plt.xlabel("Followers")
plt.ylabel("Viewers")
plt.show()
```

```python
RATIO_FOLLOWER_TO_NONE = 0.2
RATIO_FOLLOWER_CONVERSION = 0.50
FOLLOWERS_WATCH = 0.03
followers = [2]

for s in np.arange(1,50):
    none = (followers[-1]*FOLLOWERS_WATCH)/RATIO_FOLLOWER_TO_NONE
    followers.append(followers[-1]+none*RATIO_FOLLOWER_CONVERSION)

plt.plot(np.arange(1,np.array(followers).shape[0]+1), followers, c="r")
plt.xlabel("Stream")
plt.ylabel("Followers")
plt.show()
```
